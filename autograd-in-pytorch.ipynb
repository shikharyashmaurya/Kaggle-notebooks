```python
x=torch.tensor(6.7)
y=torch.tensor(0.0)
```

```python
w=torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
```

```python
z=w*x+b
```

```python
ypred=torch.sigmoid(z)
```

```python
loss=torch.nn.functional.binary_cross_entropy(ypred,y)
```

```python
loss.backward()
```

```python
w.grad,b.grad
```
