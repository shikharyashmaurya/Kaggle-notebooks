# Converters

```python
def rename(x):
    if x=='Crime':
        return 'Murder'
    else:
        return x
```

```python
rename('Crime')
```

```python
pd.read_csv('/kaggle/input/imdb-top-1000/imdb-top-1000 (1).csv',converters={'Genre':rename})
```

# if some values are placed as nan values in data

```python
pd.read_csv('/kaggle/input/imdb-top-1000/imdb-top-1000 (1).csv',na_values=['Crime'])
```
