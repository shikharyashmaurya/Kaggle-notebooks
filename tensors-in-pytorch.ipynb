you have a tensor d with shape (2, 20) and you call d.squeeze(0), it will attempt to remove the dimension at index 0 if that dimension has a size of 1. However, in this case, the dimension at index 0 has a size of 2, not 1, so calling d.squeeze(0) will not change the shape of the tensor

```python
import time
size=10000

matcpu1=torch.randn(size,size)
matcpu2=torch.randn(size,size)

starttime=time.time()
rescpu=torch.matmul(matcpu1,matcpu2)
cputime=time.time()-starttime

print(f'cpu : {cputime}')

matgpu1=matcpu1.to('cuda')
matgpu2=matcpu2.to('cuda')

starttime=time.time()
resgpu=torch.matmul(matgpu1,matgpu2)
torch.cuda.synchronize()
gputime=time.time()-starttime

print(f'cpu : {gputime}')

print(f'cpu/gpu : {cputime/gputime}')
```
